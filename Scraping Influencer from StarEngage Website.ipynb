## Step 8 : Execute

```python
sel=Selector(text=driver.page_source)
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
time.sleep(4)

for i in range(55, 100):
    
    #Name
    try:
        xname = '//*[@id="content"]/div[2]/div/div/table/tbody/tr[' + str(i+1) + ']/td[3]/text()'
        name = sel.xpath(xname).extract_first()
    except:        name = "None"

        
    #profile url
    try:
        xurl = '//*[@id="content"]/div[2]/div/div/table/tbody/tr['+ str(i+1) + ']/td[3]/a/text()'
        part_url = sel.xpath(xurl).extract_first()
        url = 'https://www.instagram.com/' + part_url[1:] + '/'
    except:
        url = "#"
    
    xengage = '//*[@id="content"]/div[2]/div/div/table/tbody/tr[' +str(i+1)+ ']/td[7]/text()'
    engagement = sel.xpath(xengage).extract_first()
    
    #go to details page of the website
    try:
        xdetails = '//*[@id="content"]/div[2]/div/div/table/tbody/tr[' +str(i+1)+ ']/td[3]/a'
        element = driver.find_element_by_xpath(xdetails)
        element.click()
    except:
        driver.execute_script("window.scrollBy(0,100);")
        xdetails = '//*[@id="content"]/div[2]/div/div/table/tbody/tr[' +str(i+1)+ ']/td[3]/a'
        element = driver.find_element_by_xpath(xdetails)
        element.click()
        
#     n = 6 + random.randint(0,5)
#     time.sleep(n)
    
    #define tabs as parent and child
    prnt = driver.window_handles[0]
    chld = driver.window_handles[1]
    
    #switch control
    driver.switch_to.window(chld)
    sel2 = Selector(text=driver.page_source)
    n = 2 + random.randint(0,3)
    time.sleep(n)
    
    #get bio
    try:
        bio = sel2.xpath('//div[@class="text-secondary small text-justify"]/text()').extract_first()
    except:
        bio = "None"
    
    #get followers and convert to number
    try:
        follower_s = sel2.xpath('//div[@class="h3 m-0 text-number pl-4 ml-3 text-light"]/text()').extract_first()
    except:
        follower_s = "None"
    
    #get posts
    try:
        post_s = sel2.xpath('//*[@id="content"]/div[2]/div/div[2]/div[2]/div[5]/div/div/div/text()').extract_first()
    except:
        post_s = "None"
    
    #switch control
    driver.close()
    driver.switch_to.window(prnt)
    
    #save to dataframe
    new_row={'Name':str(name), 'Followers':follower_s, 'Posts':post_s, 'Bio':str(bio), 'URL':str(url), 'Engagement':str(engagement)}
    data=data.append(new_row, ignore_index = True)
```
